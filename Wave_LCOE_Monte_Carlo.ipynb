```python
import numpy as np
from scipy.interpolate import interp1d
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def LR2b(LR):
    '''converts learning rate to "rate of cost reduction"'''
    return np.log2(1-LR)
```

```python
def LCOE_hist(mu,sigma_high,sigma_low,trials):
    LCOE_low = np.random.normal(mu,sigma_high,trials) # distribution using high-side sigma (low side will be ignored)
    LCOE_high = np.random.normal(mu,sigma_low,trials) # distribution using low-side sigma (high side will be ignored)
    
    highIfOne = np.random.randint(2,trials) #generate list of 0s an 1s, to be used as an index vector
    LCOE = LCOE_low #use only low-side values first
    LCOE[highIfOne] = LCOE_high[highIfOne] #then randomly substitute high-side values
    
    return LCOE
    
```

```python
#### Section 2 ####
# Figure 2: Offshore Wind Adoption Curve and Learning Rates
# Use offshore wind adoption rate
#wind_adoption = np.genfromtxt('offshore_wind_adoption.csv',delimiter=',',skip_header=1)


wind_adoption = pd.read_csv('offshore_wind_adoption.csv')
wind_adoption = wind_adoption.as_matrix()
Year = wind_adoption[:,0]
wind_InstalledMW = wind_adoption[:,1]


# Calculate Compound Annual Growth Rate for Offshore Wind
startMW = np.float64(wind_InstalledMW[0])
endMW = wind_InstalledMW[-1]
startYear = Year[0]
endYear = Year[-1]
cagr = ((endMW/startMW)**(1.0/(endYear-startYear)))-1


n_years = np.arange((Year[-1]-Year[0])+1)
cagr_coeff = np.zeros(len(n_years))
for i,y in enumerate(n_years):
    cagr_coeff[i] = (1+cagr)**i
cagr_est = cagr_coeff*wind_InstalledMW[0]


# Resample cagr-estimated offshore wind adoption to get smooth adoption curve for wave energy
f_cagr = interp1d(n_years,cagr_est)

# Plot 
fig,ax = plt.subplots()
plt.plot(Year,wind_InstalledMW/1000,linewidth=6,label='Cumulative Deployments [EWEA 2011,GWEC 2015]')
plt.plot(n_years+Year[0],cagr_est/1000,color='m',linewidth=4,label = 'Smoothed Adoption Profile using \nCompound Annual Growth Rate, 38.1%')
#plt.plot(years,f_cagr(years)/1000,'ro')    #only for checking f_cagr fit function
plt.ylabel('Cumulative Installed Capacity, GW')
plt.title('Fig 2a: Offshore Wind Deployment History')
plt.legend(fontsize=8,loc='upper left')
plt.grid()
plt.show()
```

```python
## Monte Carlo on Learning Rates
## LCOE as a function of installed capacity

#Initial Installed Capacity at 2016
#10 #Based on JRC 2016 page 22 "The expected wave energy installed capacity should have reached 370 MW in 2015 according to NREAPs [Beurskens & Hekkenberg 2011], while it has not reached 10 MW so far." 
#26.862 #Based on Table 4 in JRC 2016 
initial_installed_capacity = 10 #This fits the best, but may not be justifiable

#max_capacity = 100000 #in MW; 100000 = 100 GW
#start_capacity = np.array([initial_installed_capacity]) # first value is current installed MWs
#nearterm_capacity = np.arange(10,100,10)
#future_capacity = np.arange(100,max_capacity+100,100)  # walk in 100MW steps to 100GW
#installed_capacity = np.concatenate((start_capacity,nearterm_capacity,future_capacity))

# installed capacity in logspace
installed_capacity = np.logspace(1,5,100)

#Generate random learning rate according to offshore wind learning rates
# Learning rate sources: Rubin 2015, Wiser 2016, SI Ocean 2013
learning_rate_mean = 0.12
learning_rate_low = 0.05
learning_rate_high = 0.19
learning_rate_standard_dev = (learning_rate_high - learning_rate_low)/4 #2 standard deviations on each side of the mean, so 4 total

LCOE_trajectory = []
case = 0
LCOE_matrix = np.zeros((len(LCOE),len(installed_capacity)))

LR_matrix = np.zeros((total_trials,1))


# Calculate LCOE trajectories based on initial LCOE and a Learning Rate
for i,iLCOE in enumerate(LCOE):
    LR = np.random.normal(learning_rate_mean,learning_rate_standard_dev,1)
    LR_matrix[i] = LR
    LCOE_matrix[i,:] = iLCOE*(installed_capacity/initial_installed_capacity)**LR2b(abs(LR))
    

# Calculate and plot mean and +-2stdev 
moving_mean = np.mean(LCOE_matrix,axis=0) #mean of each row (0th axis)
moving_2stdev_low = np.percentile(LCOE_matrix,2.5,axis=0)
moving_2stdev_high = np.percentile(LCOE_matrix,97.5,axis=0)

#Calculate best-fit line for mean, +-2stdev as a function of installed capacity 
f_moving_mean = interp1d(installed_capacity,moving_mean)
f_moving_2stdev_low = interp1d(installed_capacity,moving_2stdev_low)
f_moving_2stdev_high = interp1d(installed_capacity,moving_2stdev_high)

# Need polyfit line for extrapolation
z_moving_mean = np.polyfit(installed_capacity,moving_mean,3)
p_moving_mean = np.poly1d(z_moving_mean)

z_moving_2stdev_low = np.polyfit(installed_capacity,moving_mean,3)
p_moving_2stdev_low = np.poly1d(z_moving_2stdev_low)

z_moving_2stdev_high = np.polyfit(installed_capacity,moving_mean,3)
p_moving_2stdev_high = np.poly1d(z_moving_2stdev_high)
```

```python
# Figure 2b
# Calculate and plot histogram of learning rates

n, bin_edges = np.histogram(abs(LR_matrix), 30)
bin_probability = n/float(n.sum()) # Normalize it, so that every bins value gives the probability of that bin
bin_middles = (bin_edges[1:]+bin_edges[:-1])/2. # Get the mid points of every bin
bin_width = bin_edges[1]-bin_edges[0] # Compute the bin-width
plt.bar(bin_middles, bin_probability, width=bin_width, facecolor='orange', alpha = 0.6,label='UKERC 2014 + IEA 2015 + JRC 2016 "Good Resource"')
plt.xlabel('Applied Learning Rate')
plt.ylabel('Probability')
plt.title('Figure 2b: Monte Carlo Learning Rate Distribution')
plt.show()
```
